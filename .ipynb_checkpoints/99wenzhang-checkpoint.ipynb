```python
# -*- coding: utf-8 -*-
# 大数据心理学的python入门
# 中国科学院心理研究所计算网络心理学实验室(ccpl.psych.ac.cn)

#获取99文章网中的网页内容并保存

import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin


def get_HTML_source(url):
    try:
        r = requests.get(url, timeout = 60)
        r.encoding = r.apparent_encoding  # 解决中文乱码
        r.raise_for_status()
        return r.text
    except:
        return ""

```

```python
#从解析出来的html，抽取符合搜索模式的第一个条目的内容
def get_first_item(srcSoup, regExp):
    FIContent = ""
    srcParas = srcSoup.select(regExp)
    if(len(srcParas) > 0):
        FIContent = srcParas[0].get_text()
    
    return FIContent
    
def get_all_items(srcSoup, regExp):
    FIContent = ""
    srcParas = srcSoup.select(regExp)
    for para in srcParas:
        if len(para) > 0:
            FIContent += para.get_text()
            
    return FIContent
```

```python
#下载html源文件，进行解析，抽取出其中需要信息
def extract_content(url):
    print(url)
    
    #下载html源文件
    htmlContent = get_HTML_source(url)
    if(htmlContent == ''):
        return
    
    #将html装载并解析
    soup = BeautifulSoup(htmlContent, "html.parser")

    #抽取粗标题
    title = get_first_item(soup, ".hl_c_title h2")
    print("TITLE:")
    print(title)
    print()
    
    #抽取出内容
    content = get_all_items(soup, ".hl_body")
    #print("CONTENT:")
    #print(content)
    
    #把下载的文件，保存在本地
    if(len(content) > 500):
        global save_folder
        global save_idx
        
        tzgg_file = save_folder + str(save_idx) + '.txt'
        dstfp = open(tzgg_file,'w',encoding='utf8')
        
        #dstfp.write(url)
        #dstfp.write('\n')
        
        dstfp.write(title)
        dstfp.write('\n')
        
        dstfp.write(content)
        
        dstfp.close()
        
        save_idx += 1

    return
```

```python
#获得当前网页中的列表
def get_99wz_list(url):
    html = get_HTML_source(url)
    if(html == ''):
        return
        
    soup = BeautifulSoup(html, "html.parser")

    paras = soup.select("h3 a")
    for para in paras:
        if len(para) > 0:
            rUrl = para.attrs['href']
            if(rUrl != ''):
                aUrl = urljoin(url, rUrl)
                #print(aUrl)
                
                #只处理html的网页
                if(aUrl.find('.html') != -1) or (aUrl.find('.htm') != -1):
                    extract_content(aUrl)
    return
```

```python
# 得到下一页的网址
def get_next_url(gnpUrl):
    gnpHtml = get_HTML_source(gnpUrl)
    
    gnpSoup = BeautifulSoup(gnpHtml, "html.parser")
    paras = gnpSoup.select(".page-nav a")
    for para in paras:
        if (len(para) > 0) and (para.get_text()=='下一页'):
            rUrl = para.attrs['href']
            if(rUrl != ''):
                aUrl = urljoin(gnpUrl, rUrl)
                return aUrl

    return ''

save_folder = '../data/99wz/'
save_idx = 100

if(not os.path.exists(save_folder)):
    os.makedirs(save_folder)
```

```python
# 获得“下一页”获得所有的网页
myurl = "http://www.99wenzhangwang.com/article/list_41_1.html"
while(myurl != ''):
    print(myurl)
    get_99wz_list(myurl)
    
    #我们只获取部分数据，演示用
    if(save_idx > 120):
        break
    
    myurl = get_next_url(myurl)
```
